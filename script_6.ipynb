### 2. Rimozione Metadati e Colonne Non Predittive
Rimuoviamo:
 - Metadati editoriali (pagina, licenza, immagini)
 - Dati di Lore non statistici (environments)

```python
cols_to_drop = [
    'page_no', 
    'img_main', 
    'document__slug', 
    'document__title', 
    'document__license_url', 
    'document__url',
    'environments'  # Aggiunto qui
]

# Intersezione per rimuovere solo quelle presenti
existing_cols = [c for c in cols_to_drop if c in df.columns]

if existing_cols:
    df = df.drop(columns=existing_cols)
    print(f"‚úÖ Colonne rimosse: {existing_cols}")
else:
    print("‚ö†Ô∏è Nessuna colonna da rimuovere trovata.")
```

### 3. Salvataggio
