### Import

```python
import pandas as pd
```

### Carica il csv

```python
file_path = './data/monsters.csv'  # Assicurati di specificare il percorso corretto
df = pd.read_csv(file_path)
```

```python
# Elimina le colonne specificate
columns_to_drop = ['speed.notes', 'speed.bur.']

# Rimuove le colonne dal DataFrame
df_cleaned = df.drop(columns=columns_to_drop)
```

### Salva il file aggiornato con le colonne eliminate

```python

output_file_path = './data/monsters.csv'  # Il nome del file risultante
df_cleaned.to_csv(output_file_path, index=False)

print(f"File salvato come {output_file_path}")

```
